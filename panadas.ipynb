```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Creating a dataset for Ravi's academic performance
data = {
    'Test': ['Test 1', 'Test 2', 'Test 3', 'Test 4'],
    'DSP': [75, 80, 85, 90],
    'DLIC': [70, 78, 82, 88],
    'AEC': [72, 76, 80, 85]
}
df = pd.DataFrame(data)
# Line graph
plt.figure(figsize=(8, 5))
plt.plot(df['Test'], df['DSP'], marker='o', label='DSP')
plt.plot(df['Test'], df['DLIC'], marker='s', label='DLIC')
plt.plot(df['Test'], df['AEC'], marker='^', label='AEC')
plt.title("Ravi's Academic Performance Across Tests")
plt.xlabel("Test")
plt.ylabel("Marks")
plt.legend()
plt.grid(True)
plt.show()
#bar graph
plt.figure(figsize=(8, 5))
bar_width = 0.25
x = np.arange(len(df['Test']))
plt.bar(x - bar_width, df['DSP'], width=bar_width, label='DSP', color='blue')
plt.bar(x, df['DLIC'], width=bar_width, label='DLIC', color='green')
plt.bar(x + bar_width, df['AEC'], width=bar_width, label='AEC', color='orange')
plt.xticks(x, df['Test'])
plt.title("Ravi's Marks Comparison by Test")
plt.xlabel("Test")
plt.ylabel("Marks")
plt.legend()
plt.show()
# DSP vs DLIC marks
plt.figure(figsize=(8, 5))
plt.scatter(df['DSP'], df['DLIC'], color='purple', s=100)
plt.title("Ravi's DSP vs DLIC Marks")
plt.xlabel("DSP Marks")
plt.ylabel("DLIC Marks")
plt.grid(True)
plt.show()

```
